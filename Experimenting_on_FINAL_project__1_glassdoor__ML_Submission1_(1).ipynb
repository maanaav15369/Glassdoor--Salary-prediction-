# **Project Summary -**

This project focuses on developing a Machine Learning system to predict salaries based on the Glassdoor Jobs dataset. The primary objective is to analyze job attributes and predict average salaries for various roles. We followed a standard data science process: data collection, cleaning, analysis, model building, testing, and deployment.

The dataset included detailed information such as company ratings, salary estimates, company size, industry, and required technical skills. During data cleaning, we addressed missing values, duplicate records, and irrelevant features. Salary estimates were processed to extract numerical values for minimum, maximum, and average salaries. Additionally, we engineered new features like 'company age' and skill indicators for Python, SQL, Excel, and Spark to enhance prediction accuracy.

Text from job descriptions was processed by lowercasing, removing punctuation, expanding contractions, removing URLs, digits, and stopwords, and then tokenizing, stemming, and vectorizing it using TF-IDF. Features were also standardized to optimize model performance.

We experimented with three machine learning models: Linear Regression, Decision Tree Regressor, and Random Forest Regressor. Key factors influencing salary prediction were identified as company rating, company age, and technical skills such as Python and SQL.

Ultimately, the model was saved and successfully tested on new data, providing a salary estimate of $125,994.64. This project demonstrates the utility of machine learning in supporting salary estimation, recruitment planning, and compensation benchmarking.
````python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

````

````python
#using functions to improve the code quality
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred, model_name):

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    print(f"\n{model_name} Evaluation Metrics")
    print("-" * 35)
    print(f"MAE  : {mae:.2f}")
    print(f"MSE  : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"R²   : {r2:.4f}")

    return mae, mse, rmse, r2
````

````python
#creating a plotting function
import matplotlib.pyplot as plt

def plot_metrics(mae, mse, rmse, r2, model_name):

    metrics = ["MAE", "MSE", "RMSE", "R²"]
    scores = [mae, mse, rmse, r2]

    plt.figure(figsize=(8,5))
    plt.bar(metrics, scores)

    plt.title(f"{model_name} Evaluation Metrics")
    plt.xlabel("Metrics")
    plt.ylabel("Score")

    plt.show()
````

````python
#converting the Salary Estimate into Numeric value
df['Salary Estimate'] = df['Salary Estimate'].str.replace(
    '(Glassdoor est.)','',regex=False
)
df['Salary Estimate'] = df['Salary Estimate'].str.replace(
    'Employer Provided Salary:', '', regex=False
).str.strip()

# Filter out rows where 'Salary Estimate' is '-1' to avoid ValueError
df = df[df['Salary Estimate'] != '-1']

def get_min_salary(salary_str):
    salary_str = salary_str.lower().replace('$', '').strip()
    multiplier = 1
    is_hourly = False

    if 'per hour' in salary_str:
        is_hourly = True
        salary_str = salary_str.replace('per hour', '').strip()

    if 'k' in salary_str:
        multiplier = 1000
        salary_str = salary_str.replace('k', '').strip()

    try:
        if '-' in salary_str:
            min_val = float(salary_str.split('-')[0])
        else:
            min_val = float(salary_str)

        annual_salary = min_val * multiplier
        if is_hourly:
            annual_salary *= 2080  # Assuming 40 hours/week * 52 weeks/year
        return int(annual_salary)
    except ValueError:
        return None # Or handle as appropriate

def get_max_salary(salary_str):
    salary_str = salary_str.lower().replace('$', '').strip()
    multiplier = 1
    is_hourly = False

    if 'per hour' in salary_str:
        is_hourly = True
        salary_str = salary_str.replace('per hour', '').strip()

    if 'k' in salary_str:
        multiplier = 1000
        salary_str = salary_str.replace('k', '').strip()

    try:
        if '-' in salary_str:
            max_val = float(salary_str.split('-')[1].split('(')[0].strip())
        else:
            max_val = float(salary_str)

        annual_salary = max_val * multiplier
        if is_hourly:
            annual_salary *= 2080
        return int(annual_salary)
    except ValueError:
        return None # Or handle as appropriate

df['min_salary'] = df['Salary Estimate'].apply(get_min_salary)
df['max_salary'] = df['Salary Estimate'].apply(get_max_salary)

# Drop rows where salary conversion failed (if any None values were returned)
df.dropna(subset=['min_salary', 'max_salary'], inplace=True)

df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2
````

````python
df['Salary Estimate'] = df['Salary Estimate'].str.replace(
    '(Glassdoor est.)','',regex=False
) # Remove 'Glassdoor est.'
df['Salary Estimate'] = df['Salary Estimate'].str.replace(
    'Employer Provided Salary:', '', regex=False
).str.strip() # Remove employer provided salary text

# Filter out rows where 'Salary Estimate' is '-1' and create an explicit copy
df = df[df['Salary Estimate'] != '-1'].copy() # Exclude invalid salary estimates and create a copy

def get_min_salary(salary_str): # Extract minimum salary from string
    salary_str = salary_str.lower().replace('$', '').strip() # Clean salary string
    multiplier = 1 # Initialize multiplier
    is_hourly = False # Flag for hourly rates

    if 'per hour' in salary_str: # Check for hourly rates
        is_hourly = True # Set hourly flag
        salary_str = salary_str.replace('per hour', '').strip() # Remove 'per hour'

    if 'k' in salary_str: # Check for 'k' (thousands)
        multiplier = 1000 # Set multiplier to 1000
        salary_str = salary_str.replace('k', '').strip() # Remove 'k'

    try: # Handle conversion errors
        if '-' in salary_str: # Check for salary range
            min_val = float(salary_str.split('-')[0]) # Extract min value
        else: # Single value salary
            min_val = float(salary_str) # Convert to float

        annual_salary = min_val * multiplier # Calculate annual salary
        if is_hourly: # Adjust for hourly rates
            annual_salary *= 2080  # Convert to annual (40h/week * 52weeks)
        return int(annual_salary) # Return integer annual salary
    except ValueError: # If conversion fails
        return None # Return None

def get_max_salary(salary_str): # Extract maximum salary from string
    salary_str = salary_str.lower().replace('$', '').strip() # Clean salary string
    multiplier = 1 # Initialize multiplier
    is_hourly = False # Flag for hourly rates

    if 'per hour' in salary_str: # Check for hourly rates
        is_hourly = True # Set hourly flag
        salary_str = salary_str.replace('per hour', '').strip() # Remove 'per hour'

    if 'k' in salary_str: # Check for 'k' (thousands)
        multiplier = 1000 # Set multiplier to 1000
        salary_str = salary_str.replace('k', '').strip() # Remove 'k'

    try: # Handle conversion errors
        if '-' in salary_str: # Check for salary range
            max_val = float(salary_str.split('-')[1].split('(')[0].strip()) # Extract max value
        else: # Single value salary
            max_val = float(salary_str) # Convert to float

        annual_salary = max_val * multiplier # Calculate annual salary
        if is_hourly: # Adjust for hourly rates
            annual_salary *= 2080 # Convert to annual
        return int(annual_salary) # Return integer annual salary
    except ValueError: # If conversion fails
        return None # Return None

df['min_salary'] = df['Salary Estimate'].apply(get_min_salary) # Create min_salary column
df['max_salary'] = df['Salary Estimate'].apply(get_max_salary) # Create max_salary column

# Drop rows where salary conversion failed
df.dropna(subset=['min_salary', 'max_salary'], inplace=True) # Remove rows with null min/max salaries

df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2 # Calculate average salary
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Three hypotheses were tested using statistical methods. Pearson Correlation was used to test the relationship between company rating and salary, ANOVA was used to compare salaries across company sizes, and an Independent T-Test was used to compare salaries of Python and non-Python jobs. A p-value less than 0.05 indicates a statistically significant relationship, leading to rejection of the null hypothesis. These tests help validate insights obtained from the visualizations.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 1

Statement: Companies with higher ratings offer higher average salaries.

Null Hypothesis (H0): Company rating has no significant effect on average salary.

Alternative Hypothesis (H1): Higher company ratings are associated with higher average salaries.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['Rating'], df['avg_salary'])

print("Correlation:", corr)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
````

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value for Hypothesis 1, I used the Pearson Correlation Test (pearsonr) because both Company Rating and Average Salary are numerical variables. The test measures the strength of their relationship and determines whether the correlation is statistically significant using the p-value.

##### Why did you choose the specific statistical test?

Both variables (Rating and Average Salary) are numerical.

Pearson Correlation measures the strength and direction of the linear relationship between two continuous variables.

It also provides a p-value to determine whether the observed correlation is statistically significant.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 2

Statement: Larger companies offer higher salaries than smaller companies.

H0: No significant difference in average salary among company sizes.

H1: Significant difference exists in average salary among company sizes.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

groups = [group['avg_salary'].dropna()
          for name, group in df.groupby('Size')]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
````

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value for Hypothesis 2, I used the One-Way ANOVA Test (f_oneway) because Company Size is a categorical variable with multiple groups and Average Salary is a numerical variable. ANOVA determines whether there is a statistically significant difference in average salaries among different company size categories.

##### Why did you choose the specific statistical test?

Company Size is a categorical variable with multiple groups (Small, Medium, Large, etc.).

Average Salary is a numerical variable.

ANOVA is used to compare the means of more than two groups and determine whether at least one group differs significantly from the others.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 3

Statement: Jobs requiring Python have higher salaries than jobs not requiring Python.

H0: No significant salary difference exists between Python and non-Python jobs.

H1: Python-related jobs have significantly higher salaries.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

python_jobs = df[df['python_skill'] == 1]['avg_salary']
non_python_jobs = df[df['python_skill'] == 0]['avg_salary']

t_stat, p_value = ttest_ind(
    python_jobs,
    non_python_jobs,
    equal_var=False
)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
````

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value for Hypothesis 3, I used an Independent Two-Sample T-Test (Welch's T-Test) because I was comparing the mean salaries of two independent groups: jobs requiring Python and jobs not requiring Python. The test determines whether the difference in average salaries between the two groups is statistically significant.

##### Why did you choose the specific statistical test?

We are comparing the average salary of two independent groups:

Jobs requiring Python

Jobs not requiring Python

The dependent variable (Average Salary) is numerical.

The independent variable (Python Skill) has two categories (Yes/No).

Welch's T-Test (equal_var=False) is preferred because the two groups may have different variances.

````python
# Remove URLs & Remove words and digits contain digits
import re

# Remove URLs
df['Job Description'] = df['Job Description'].apply(
    lambda x: re.sub(r'http\S+|www\S+|https\S+', '', str(x))
)

# Remove words containing digits
df['Job Description'] = df['Job Description'].apply(
    lambda x: re.sub(r'\w*\d\w*', '', x)
)
````

````python
# Remove Stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df['Job Description'] = df['Job Description'].apply(
    lambda x: ' '.join(
        word for word in str(x).split()
        if word not in stop_words
    )
)
````

### ML Model - 3

````python
# ML Model - 3 Implementation

# Fit the Algorithm
from xgboost import XGBRegressor
import pandas as pd
import numpy as np

xgb_model = XGBRegressor(
    random_state=42,
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6
)

# Explicitly ensure 'Rating' and 'company_age' are standard numpy float64 dtype
# This addresses cases where pd.NA might cause the column to be interpreted as 'object'
# by XGBoost, even if the pandas dtype appears numeric.
for col in ['Rating', 'company_age']:
    # Step 1: Use pd.to_numeric to robustly convert, coercing errors to NaN
    # This handles any mixed types or non-numeric strings and converts pd.NA to np.nan
    X_train_cleaned.loc[:, col] = pd.to_numeric(X_train_cleaned[col], errors='coerce')
    X_test_cleaned.loc[:, col] = pd.to_numeric(X_test_cleaned[col], errors='coerce')

    # Step 2: Fill any NaNs that might have been introduced by coerce or were initially present.
    # Though `dropna` should have cleared most, this adds robustness.
    # Using median for imputation as it's less sensitive to outliers than mean.
    median_train = X_train_cleaned[col].median()
    X_train_cleaned.loc[:, col] = X_train_cleaned[col].fillna(median_train)
    median_test = X_test_cleaned[col].median()
    X_test_cleaned.loc[:, col] = X_test_cleaned[col].fillna(median_test)

    # Step 3: Explicitly cast to 'float64' numpy dtype to guarantee compatibility with XGBoost
    X_train_cleaned.loc[:, col] = X_train_cleaned[col].astype('float64')
    X_test_cleaned.loc[:, col] = X_test_cleaned[col].astype('float64')

# Convert DataFrames to NumPy arrays before passing to XGBoost to bypass any potential Pandas dtype issues.
# This is the most robust way to ensure XGBoost receives purely numerical data.
X_train_cleaned_np = X_train_cleaned.to_numpy()
y_train_cleaned_np = y_train_cleaned.to_numpy()

# Fit the model with the NumPy arrays
xgb_model.fit(X_train_cleaned_np, y_train_cleaned_np)
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
import pandas as pd
import numpy as np

# Ensure 'Rating' and 'company_age' are robustly converted to numeric and NaNs handled
# This step is repeated here to guarantee consistency for GridSearchCV
for col in ['Rating', 'company_age']:
    # Convert to numeric, coercing errors to NaN
    X_train_cleaned.loc[:, col] = pd.to_numeric(X_train_cleaned[col], errors='coerce')
    X_test_cleaned.loc[:, col] = pd.to_numeric(X_test_cleaned[col], errors='coerce')

    # Impute NaNs with the median from the training set
    median_train = X_train_cleaned[col].median()
    X_train_cleaned.loc[:, col] = X_train_cleaned[col].fillna(median_train)

    # Use the median from the training set to impute NaNs in the test set to prevent data leakage
    X_test_cleaned.loc[:, col] = X_test_cleaned[col].fillna(median_train) # Use train median for test set imputation

    # Explicitly cast to float64
    X_train_cleaned.loc[:, col] = X_train_cleaned[col].astype('float64')
    X_test_cleaned.loc[:, col] = X_test_cleaned[col].astype('float64')

# Define Model
xgb = XGBRegressor(
    objective='reg:squarederror',
    random_state=42
)

# Hyperparameter Grid
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

# Convert DataFrames to NumPy arrays before passing to GridSearchCV
# This is the most robust way to ensure XGBoost receives purely numerical data.
X_train_cleaned_np = X_train_cleaned.to_numpy()
y_train_cleaned_np = y_train_cleaned.to_numpy()
X_test_cleaned_np = X_test_cleaned.to_numpy()

# Grid Search
grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)
# Fit the Algorithm
grid_search.fit(X_train_cleaned_np, y_train_cleaned_np)

print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)

# Predict on the model
# Best Model
best_xgb = grid_search.best_estimator_

# Prediction using the NumPy array version of the test set
y_pred_xgb = best_xgb.predict(X_test_cleaned_np)
````
